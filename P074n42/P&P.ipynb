```python
import plotly.graph_objects as go
import numpy as np
from scipy import interpolate

path = "data/p&p/"
files = ['01_pump_23mW_probe_3mW', '02_pump_17mW_probe_3mW', '03_pump_12mW_probe_3mW', '04_pump_7mW_probe_3mW', '05_pump_1mW_probe_3mW_CHoppS']
pumps = {
    '01_pump_23mW_probe_3mW': "23 mW",
    '02_pump_17mW_probe_3mW': "17 mW",
    '03_pump_12mW_probe_3mW': "12 mW",
    '04_pump_7mW_probe_3mW': "7 mW",
    '05_pump_1mW_probe_3mW_CHoppS': '1 mW'
}
chopps_file = '05_pump_1mW_probe_3mW_CHoppS'
A0_data = np.loadtxt(path+chopps_file+".dat", skiprows=1)
A0 = np.average(A0_data[:,3])

skok = []
fig = go.Figure()
for curve in files:
    data = np.loadtxt(path+curve+".dat", skiprows=1)
    starts = np.argwhere(data[:,0] == np.amin(data[:,0]))
    starts = [i[0] for i in starts]
    cycles = np.split(data, starts, axis=0)
    averaged = np.mean(cycles[1:], axis=0)
    data_x = averaged[:,0]/1000-2321.997
    data_y = averaged[:,3]
    #data_y = (data_y-np.average(data_y[(data_x>2.26)*(data_x<2.31)]))*100
    data_y = (data_y)*1000
    fig.add_trace(go.Scatter(x=data_x, y=data_y,
                        mode='markers+lines',
                    name=pumps[curve]))
    data_y = (averaged[:,3]-A0)*1000
    S0 = np.average(data_y[(data_x>5)])
    S_t = np.average(data_y[(data_x<-5)*(data_x>-65)])
    tau = 0.6/np.log(S_t/S0)
    skok.append(S0-S_t)
    print('{}: Τ = {} ms'.format(pumps[curve], round(tau, 2)))
fig.update_layout(
    xaxis_title="Time delay (ps)",
    yaxis_title="Signal (mV)",
    legend_title="Probe:<br>@ 3 mW<br><br>Pump:",
)

fig.show()
```
