```python
import plotly.graph_objects as go
import numpy as np
from scipy import interpolate

path = r"D:/P074n35B2/data/"
files = ['Ba1', 'Ba2', 'Ba3', 'Ba4', 'Ba5', 'Ba6']

fig = go.Figure()

for curve in files:
    data = np.loadtxt(path+curve+".dat", skiprows=1)
    fig.add_trace(go.Scatter(x=data[:,0]/1000000, y=1000*data[:,3],
                        mode='markers+lines',
                    name=curve))

fig.show()
```

```python
path = r"D:/P074n42/data/p&p/"
files = ['01_pump_23mW_probe_3mW', '02_pump_23mW_probe_3mW', 'Bb3', 'Bb4','Bb5', 'Bb6', 'Bb6a']
#files = ['Bc1', 'Bc2', 'Bc3', 'Bc4', 'Bc5', 'Bc6', 'Bc7', 'Bc8', 'Bc9break']#, 'Bb4', 'Bb5', 'Bb6']

fig = go.Figure()

for curve in files:
    data = np.loadtxt(path+curve+".dat", skiprows=1)
    starts = np.argwhere(data[:,0] == np.amin(data[:,0]))
    starts = [i[0] for i in starts]
    cycles = np.split(data, starts, axis=0)
    averaged = np.mean(cycles[1:], axis=0)
    
    fig.add_trace(go.Scatter(x=averaged[:,0]/1000000, y=1000*averaged[:,3],
                        mode='markers+lines',
                    name=curve))

fig.show()
```
