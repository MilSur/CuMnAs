```python
#@title Závislost na opakovací frekvenci

#@markdown - Pharos @ 190kHz, volba DR urcuje opakovaci frekveci
#@markdown - device P074n42H4 (10 um), pocatecni odpor 37.7 Ω
#@markdown - **probe:** 1030nm, w_x ~ 18 um, w_y ~ 26 um, zacina switchovat kolem 23 mW (I_th ~ 20 mW), @ 29 mW signal kolem 0.5 % **-> nastaven fixne na 15 mW**
#@markdown - **pump**: 1030nm, w_x ~ 17 um, w_y ~ 29 um, zacina switchovat kolem 40 mW (I_th ~ 35 mW), @ 47 mW signal kolem 0.5 % **-> nastaven fixne na 15 mW**
#@markdown - čtení lockinem: 4p měření, kde algebrator měří potenciálový rozdíl mezi A(4) - B(5), zesílení 200x; Lockin: 300ms, fáze -159deg

import plotly.graph_objects as go
import numpy as np
from scipy import interpolate

path = "/content/CuMnAs/data/P074n42/korelace/H4 15mW pump 15mW probe/"
files = ['01_100Hz', '02_80Hz', '03_60Hz', '04_40Hz', '05_20Hz']
pumps = {
    '01_100Hz': '100 Hz',
    '02_80Hz': '80 Hz',
    '03_60Hz': '60 Hz',
    '04_40Hz': '40 Hz',
    '05_20Hz': '20 Hz',
}

fig = go.Figure()
for curve in files:
    data = np.loadtxt(path+curve+".dat", skiprows=1)
    starts = np.argwhere(data[:,0] == np.amin(data[:,0]))
    starts = [i[0] for i in starts]
    cycles = np.split(data, starts, axis=0)
    averaged = np.mean(cycles[1:], axis=0)
    data_x = averaged[:,0]/1000-2321.997
    data_y = averaged[:,3]
    #data_y = (data_y-np.average(data_y[(data_x>2.26)*(data_x<2.31)]))*100
    data_y = (data_y)*1000
    fig.add_trace(go.Scatter(x=data_x, y=data_y,
                        mode='markers+lines',
                    name=pumps[curve]))
    
fig.update_layout(
    xaxis_title="Time delay (ps)",
    yaxis_title="Signal (mV)",
    legend_title="Both Pump & Probe:<br>@ 15 mW<br><br>Repetition Rate:",
)

fig.show()
```
