```python
#@title Intenzitní závislost

#@markdown - 100 Hz, (Pharos @ 190kHz, DR=1902)
#@markdown - device P074n42G1 (10 um), pocatecni odpor 40.1 Ω
#@markdown - **probe:** 1030nm, w_x ~ 18 um, w_y ~ 26 um, zacina switchovat kolem 23 mW (I_th ~ 20 mW), @ 29 mW signal kolem 0.5 % **-> nastaven fixne na 15 mW pro korelace**
#@markdown - **pump**: 1030nm, w_x ~ 17 um, w_y ~ 29 um, zacina switchovat kolem 40 mW (I_th ~ 35 mW), @ 47 mW signal kolem 0.5 % **-> na 25 mW by mel v prekryvu switchovat**
#@markdown - mereno pro fixni probe a promenny pumpu od 20 mW nize a pote zpet nahoru, nakonci je ale device trochu pripaleny (48.15 Ω)
#@markdown - čtení lockinem: 4p měření, kde algebrator měří potenciálový rozdíl mezi A(9) - B(8), zesílení 200x; Lockin: 300ms, fáze -159deg

# allfiles = ['01_100ms', '02_300ms_20mW', '03_15mW', '04_10mW', '05_5mW', '06_0mW', '07_18mW', '08_22mW', '08_24mW', '10_25mW', '11_25mW_fine', '12_20mW']

path = "/content/CuMnAs/data/P074n42/korelace/100 Hz/"
files = ['02_300ms_20mW', '03_15mW', '04_10mW', '05_5mW', '06_0mW', '07_18mW', '08_22mW', '08_24mW', '10_25mW', '11_25mW_fine', '12_20mW']
pumps = {
    '02_300ms_20mW': '20 mW',
    '03_15mW': '15 mW',
    '04_10mW': '10 mW',
    '05_5mW': '5 mW',
    '06_0mW': '0 mW',
    '07_18mW': '18 mW',
    '08_22mW': '22 mW',
    '08_24mW': '24 mW',
    '10_25mW': '25 mW',
    '11_25mW_fine': '25 mW',
    '12_20mW': '20 mW',
}

fig = go.Figure()
for curve in files:
    data = np.loadtxt(path+curve+".dat", skiprows=1)
    starts = np.argwhere(data[:,0] == np.amin(data[:,0]))
    starts = [i[0] for i in starts]
    cycles = np.split(data, starts, axis=0)
    averaged = np.mean(cycles[1:], axis=0)
    data_x = averaged[:,0]/1000-2321.997
    data_y = averaged[:,3]
    #data_y = (data_y-np.average(data_y[(data_x>2.26)*(data_x<2.31)]))*100
    data_y = (data_y)*1000
    fig.add_trace(go.Scatter(x=data_x, y=data_y,
                        mode='markers+lines',
                    name=pumps[curve]))
    
fig.update_layout(
    xaxis_title="Time delay (ps)",
    yaxis_title="Signal (mV)",
    legend_title="Probe:<br>@ 15 mW<br><br>Pump:",
)

fig.show()
```
